```python
 # Removing the duplicate values in the data
df.drop_duplicates(inplace=True)
#Exploring the dataset
df
```

```python
#Identifying outliers and removing them
# calculate IQR for each column
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

# identify outliers
outliers = ((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any()
outliers
# remove outliers
df_no_outliers = df.mask(outliers)

# print number of outliers removed
print(f"Number of outliers removed: {df.isna().sum().sum()}")

# drop rows with missing values
df_no_outliers.dropna(inplace=True)
```

```python
# Visualising with the help of Count Plot in Seaborn
sns.set_style('whitegrid')
sns.countplot(x='spam_nospam',data=df,palette='viridis')

# Count of spam and nonspam values in the data set
spam = df[df['spam_nospam'] == 1].count()['spam_nospam']
nonspam = df[df['spam_nospam'] == 0].count()['spam_nospam']

print(f'Spam Values: {spam}')
print(f'Nonspam Values: {nonspam}')
```

```python
#separate target values
y = df['spam_nospam'].values

#view target values
y[0:5]
```

```python
#Standardizing features
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_data=scaler.fit_transform(df)
scaled_data
```

```python
df=pd.DataFrame(data=scaled_data, columns= df.columns)
df.describe(), df
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
new_data = pca.fit_transform(scaled_data)
# This will be the new data fed to the algorithm.
principal_Df = pd.DataFrame(data = new_data
             , columns = ['principal component 1', 'principal component 2','principal component 3'])
```

```python
principal_Df
```

```python
from sklearn.model_selection import train_test_split
#spitting train and test data
XTrain, XTest, YTrain, YTest = train_test_split(principal_Df, y, test_size = 0.25, random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
# train KNN model
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(XTrain, YTrain)

# predict on test set and calculate accuracy
knn_preds = knn_model.predict(XTest)
print("KNN model accuracy:" ,knn_model.score(XTest, YTest))
```

```python
from sklearn.metrics import classification_report, confusion_matrix
class_report = classification_report(YTest, knn_preds)
print("Classification Report: \n",class_report)
conf_matrix = confusion_matrix(YTest, knn_preds)
print("Confusion Matrix: \n",conf_matrix)
```

```python
from sklearn.model_selection import GridSearchCV
#Checking for best n neighbour
#create new a knn model
knn2 = KNeighborsClassifier()

#create a dictionary of all values we want to test for n_neighbors
param_grid = {'n_neighbors': np.arange(1, 25)}

#  Use gridsearch to test all values for n_neighbors
knn_gscv = GridSearchCV(knn_model, param_grid, cv=10, scoring='accuracy', return_train_score=False,verbose=1)
#fit model to data
knn_gscv.fit(principal_Df, y)
```

```python
#check top performing n_neighbors value
knn_gscv.best_params_
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
# train Decision Tree model
dt_model = DecisionTreeClassifier(random_state=0,criterion='gini')
dt_model.fit(XTrain, YTrain)

# predict on test set and calculate accuracy
dt_preds = dt_model.predict(XTest)
dt_acc = accuracy_score(YTest, dt_preds)
print("Decision Tree model accuracy: ",dt_acc)
```

```python
#classification_report
print("Classification Report: \n",classification_report(YTest, dt_preds))

#confusion matrix
print("Confusion matrix: \n",confusion_matrix(YTest, dt_preds))
```
