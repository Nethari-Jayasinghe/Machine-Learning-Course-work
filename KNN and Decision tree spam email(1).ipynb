```python
 # Removing the duplicate values in the data
df.drop_duplicates(inplace=True)
#Exploring the dataset
df
```

```python
# calculate IQR for each column
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

# identify outliers
outliers = ((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any()
outliers
# remove outliers
df_no_outliers = df.mask(outliers)

# print number of outliers removed
print(f"Number of outliers removed: {df.isna().sum().sum()}")

# drop rows with missing values
df_no_outliers.dropna(inplace=True)
```

```python
# Visualising with the help of Count Plot in Seaborn
sns.set_style('whitegrid')
sns.countplot(x='spam_nospam',data=df,palette='viridis')

# Count of spam and nonspam values in the data set
spam = df[df['spam_nospam'] == 1].count()['spam_nospam']
nonspam = df[df['spam_nospam'] == 0].count()['spam_nospam']

print(f'Spam Values: {spam}')
print(f'Nonspam Values: {nonspam}')
```

```python
#separate target values
y = df['spam_nospam'].values

#view target values
y[0:5]
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_data=scaler.fit_transform(df)
```

```python
df=pd.DataFrame(data=scaled_data, columns= df.columns)
df.describe()
```

```python
from sklearn.decomposition import PCA
pca = PCA()
principalComponents = pca.fit_transform(scaled_data)
plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Variance (%)') #for each component
plt.title('Explained Variance')
plt.show()
```

```python
#create a dataframe with all training data except the target column
DF = df.drop(columns=['spam_nospam'])

#check that the target variable has been removed
DF.head()
```

```python
from sklearn.model_selection import train_test_split
#spitting train and test data

XTrain, XTest, YTrain, YTest = train_test_split(DF, y, test_size = 0.25, random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
# train KNN model
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(XTrain, YTrain)

# predict on test set and calculate accuracy
knn_preds = knn_model.predict(XTest)
knn_acc = accuracy_score(YTest, knn_preds)
print(f"KNN model accuracy: {knn_acc:.4f}")
```

```python
y_pred = knn_model.predict(XTest)
```

```python
#check accuracy of our model on the test data
knn_model.score(XTest, YTest)
```
